## Base line prediction: ARIMA

ARIMA, short for ‘AutoRegressive Integrated Moving Average’, is a forecasting algorithm based on the idea that the information in the past values of the time series can alone be used to predict the future values.

```python
line_plot_adj_price(df, 'Adj Close')
```

On the graph above, we can see that the stock adjusted closing price doesn't have a constant variance. Therefore, we need to take a log of it, to stabilize the variance.

```python
# remove variability of the data using log
df['Log Adj Close'] = np.log2(df['Adj Close']) 
```

```python
line_plot_adj_price(df, 'Log Adj Close')
```

### Data Pre-processing 

`1.` Use Augmented Dickey Fuller test to check if TS is stationary:

The null hypothesis of the ADF test is that the time series is non-stationary. So, if the p-value of the test is less than the significance level (0.05) then you reject the null hypothesis and infer that the time series is indeed stationary.

```python
# create a subset of the data
df_adj = df.loc[:,['Log Adj Close']]

result = adfuller(df_adj['Log Adj Close'].values)

print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])

if result[1] > 0.05:
    print('Fail to reject the null hypothesis that the TS is non-stationary (i.e TS is non-stationary)')
else:
    print('Reject the null hypothesis that the TS is non-stationary. (i.e TS is stationary)')
```

Since TS is non-stationary we need to perform differencing to convert TS to stationary.
Note, differencing will remove the trend. 

---

`2.` Find the order of differencing d.

```python
#NOTE: I am not ploting acf at lag 0 as it is always = 1.

# Original Series
fig, axes = plt.subplots(4, 1, sharex=False, figsize=(10,22))
axes[0].plot(df_adj['Log Adj Close'], linewidth=0.9)
axes[0].set_title('Original Series')
plot_acf(
    df_adj['Log Adj Close'],
    ax=axes[1],
    lags=40,
    linewidth=0.5,
    alpha=0.05,
    markersize=0.1,
    marker=","
)

# 1st Differencing
axes[2].plot(df_adj['Log Adj Close'].diff(), linewidth=0.5)
axes[2].set_title('1st Order Differencing')
plot_acf(
    df_adj['Log Adj Close'].diff().dropna(),
    ax=axes[3],
    alpha=0.05,
    lags=40,
    linewidth=0.1,
    markersize=0.1,
    marker=",",
)

plt.show()
```

The slow decrease in the original ACF as the lags increase is due to the trend. 
d = 1

----

`3.` Find the order of the AR term (p)

The required number of AR terms can be found by inspecting the Partial Autocorrelation (PACF) plot.

```python
# PACF plot of 1st differenced series
fig, axes = plt.subplots(2, 1, sharex=False, figsize=(10,8))
axes[0].plot(df_adj['Log Adj Close'].diff().dropna(), linewidth=0.5)
axes[0].set_title('1st Differencing')
axes[1].set(ylim=(-0.1,0.1))
plot_pacf(df_adj['Log Adj Close'].diff().dropna(), ax=axes[1], zero=False)

plt.show()
```

We can observe that the PACF lag 6 is quite significant since is well outside of the significance boundaries. 

---

`4.` Find the order of the MA term (q)

The ACF plot tells us how many MA terms are required to remove any autocorrelation in the stationarized series.

```python
fig, axes = plt.subplots(2, 1, sharex=False, figsize=(10,8))
axes[0].plot(df_adj['Log Adj Close'].diff(), linewidth=0.5)
axes[0].set_title('1st Differencing')
axes[1].set(ylim=(-0.1,0.1))
plot_acf(df_adj['Log Adj Close'].diff().dropna(), ax=axes[1], zero=False)

plt.show()
```

### The Durbin Watson (DW) test

The Durbin Watson (DW) statistic is a test for autocorrelation in the residuals from a statistical regression analysis. The Durbin-Watson statistic will always have a value between 0 and 4. A value of 2.0 means that there is no autocorrelation detected in the sample. Values from 0 to less than 2 indicate positive autocorrelation and values from from 2 to 4 indicate negative autocorrelation. [source](https://www.investopedia.com/terms/d/durbin-watson-statistic.asp)

```python
sm.stats.durbin_watson(ar_model.resid.values)
```

```python
fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(111)
ax = ar_model.resid.plot(ax=ax);
```

```python
# Normality of Residual test
resid = ar_model.resid
stats.normaltest(resid)
```

```python
# null hypothesis: x comes from a normal distribution
if stats.normaltest(resid).pvalue < 0.01:  
    print("Reject the null hypothesis that residuals come from a normal distribution.")
else:
    print("Fail to reject the null hypothesis that residuals come from a normal distribution.")
```

```python
fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(111)
fig = qqplot(resid, line='q', ax=ax, fit=True)
```

Note: the QQPlot about suggests heavy-tailed residuals. In other words, those points are much further from the mean than one would expect if the data generating process was actually a normal distribution.

```python
fig = plt.figure(figsize=(12,8))
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(resid.values.squeeze(), lags=40, ax=ax1)
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(resid, lags=40, ax=ax2)
```

```python
# train test split ratio
size = int(df_adj.shape[0] * 0.80)

# train test split
X = df_adj['Adj Close'].values
train, test = X[0:size], X[size:]

# initialization
history = [x for x in train]
predictions = list()

# walk-forward validation
for t in range(len(test)):
    model = ARIMA(history, order=(1, 0, 0))
    model_fit = model.fit(method='innovations_mle')
    output = model_fit.forecast()
    yhat = output[0]
    predictions.append(yhat)
    obs = test[t]
    history.append(obs)
    
# evaluate forecasts
rmse = sqrt(mean_squared_error(test, predictions))
print('Test RMSE: %.3f' % rmse)
```

```python
# plot forecasts against actual outcomes
plt.plot(test)
plt.plot(predictions, color='blue')
plt.show()
```

## Feature Engineering

```python
df['HL_PCT'] = ((df['High'] - df['Low']) / df['Adj Close'] * 100).values
df['PCT_change'] = ((df['Adj Close'] - df['Open']) / df['Open'] * 100.0).values

df_new = df[['Adj Close', 'HL_PCT', 'PCT_change']]
df_new
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor

from sklearn.linear_model import Ridge
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline

import math
import datetime

from sklearn import preprocessing
from sklearn.model_selection import train_test_split


# Drop missing value
df_new = df_new.fillna(value=-99999)

# We want to separate 1 percent of the data to forecast
forecast_out = int(math.ceil(0.01 * len(df_new)))

# Separating the label here, we want to predict the AdjClose
forecast_col = 'Adj Close'
df_new['label'] = df_new[forecast_col].shift(-forecast_out)
X = np.array(df_new.drop(['label'], 1))

# Scale the X so that everyone can have the same distribution for linear regression
X = preprocessing.scale(X)

# Finally we want to find Data Series of late X and early X (train) for model generation and evaluation
X_test = X[-forecast_out:]
X_train = X[:-forecast_out]

# Separate label and identify it as y
y = np.array(df_new['label'])
y_train = y[:-forecast_out]
y_test = np.array(df_new['Adj Close'])
y_test = y_test[-forecast_out:]

# Linear regression
clfreg = LinearRegression(n_jobs=-1)
clfreg.fit(X_train, y_train)

# Quadratic Regression 2
clfpoly2 = make_pipeline(PolynomialFeatures(2), Ridge())
clfpoly2.fit(X_train, y_train)

# Quadratic Regression 3
clfpoly3 = make_pipeline(PolynomialFeatures(3), Ridge())
clfpoly3.fit(X_train, y_train)

```

```python
# KNN Regression
clfknn = KNeighborsRegressor(n_neighbors=2)
clfknn.fit(X_train, y_train)
```

```python
confidencereg = clfreg.score(X_test, y_test)
confidencepoly2 = clfpoly2.score(X_test,y_test)
confidencepoly3 = clfpoly3.score(X_test,y_test)
confidenceknn = clfknn.score(X_test, y_test)


confidencereg, confidencepoly2, confidencepoly3, confidenceknn
```

```python
forecast_set = clfreg.predict(X_test)
df_new['Forecast'] = np.nan
```

```python
last_date = df_adj.iloc[-1].name
last_unix = last_date
next_unix = last_unix + datetime.timedelta(days=1)

for i in forecast_set:
    next_date = next_unix
    next_unix += datetime.timedelta(days=1)
    df_new.loc[next_date] = [np.nan for _ in range(len(df_new.columns)-1)]+[i]

df_new['Adj Close'].tail(500).plot()
df_new['Forecast'].tail(500).plot()

plt.legend(loc=4)
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```
